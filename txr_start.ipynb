```python
df['AR_TO_PL'] = (df['AR_TO_PL'].str.split(':').str[0].astype('int64'))*60 + (df['AR_TO_PL'].str.split(':').str[1].astype('int64'))
```

```python
df['PL_TO_RL'] = (df['PL_TO_RL'].str.split(':').str[0].astype('int64'))*60 + (df['PL_TO_RL'].str.split(':').str[1].astype('int64'))
df['RL_TO_DP'] = (df['RL_TO_DP'].str.split(':').str[0].astype('int64'))*60 + (df['RL_TO_DP'].str.split(':').str[1].astype('int64'))
df['TOTAL_TIME'] = (df['TOTAL_TIME'].str.split(':').str[0].astype('int64'))*60 + (df['TOTAL_TIME'].str.split(':').str[1].astype('int64'))
```

```python
df['RDATE'] = pd.to_datetime(df['RDATE'], format='%d-%m-%Y %H:%M')

df['ARVLTIME'] = pd.to_datetime(df['ARVLTIME'], format='%d-%m-%Y %H:%M')
df['PLCTTIME'] = pd.to_datetime(df['PLCTTIME'], format='%d-%m-%Y %H:%M')
df['RELSTIME'] = pd.to_datetime(df['RELSTIME'], format='%d-%m-%Y %H:%M')
df['DPRTTIME'] = pd.to_datetime(df['DPRTTIME'], format='%d-%m-%Y %H:%M')
```

```python
df['R_DATEOFMONTH'] = df['RDATE'].dt.date
df['R_MONTH'] = df['RDATE'].dt.to_period('M')
df['R_DAYOFWEEK'] = df['RDATE'].dt.dayofweek

df['YEAR'] = df['RDATE'].dt.isocalendar().year
df['WEEK'] = df['RDATE'].dt.isocalendar().week
df['R_WEEK'] = 'W-' + df['WEEK'].astype(str) + '-' + df['YEAR'].astype(str)
df.drop(columns=['YEAR','WEEK'], axis=1, inplace=True)

df['ARVL_H'] = df['ARVLTIME'].dt.hour
df['PLCT_H'] = df['PLCTTIME'].dt.hour
df['RELS_H'] = df['RELSTIME'].dt.hour
df['DPRT_H'] = df['DPRTTIME'].dt.hour
```
